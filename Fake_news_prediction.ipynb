```python
X=news.drop(columns='labels',axis=1)
Y=news['labels']
```

```python
port_sem=PorterStemmer()
```

```python
def stemmmer(contant):
  contant_stammed=re.sub('[^a-zA-Z]',' ',contant)
  contant_stammed=contant_stammed.lower()
  contant_stammed=contant_stammed.split()
  contant_stammed=[port_sem.stem(word) for word in contant_stammed if not word in stopwords.words('english')]
  contant_stammed=' '.join(contant_stammed)
  return contant_stammed
```

```python
news['article_content']=news['article_content'].apply(stemmmer)
```

```python
news
```

```python
X=news['article_content'].values
Y=news['labels'].values
```

```python
Y
```

VActorization

```python
vectorization=TfidfVectorizer()
vectorization.fit(X)
X=vectorization.transform(X)
```

```python
print(X)
```

Spliting data


```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

Logistic regression

```python
X=news['article_content'].values
Y=news['labels'].values
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
x_pridict=lr.predict(x_test)
accuracy_score(x_pridict,y_test)
```
